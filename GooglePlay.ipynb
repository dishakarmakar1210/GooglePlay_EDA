```python
#1Mb = 1024 Kb
def convert_size(item):
    if item[-1] == 'M':
        return float(item[:-1]) * 1024 #convert to kb
    elif item[-1] == 'k':
        return float(item[:-1])
    elif item[-1] == 'G':
        return float(item[:-1]) * 1024 * 1024 #convert to kb
    else:
        return np.nan
#apply in size column
```

```python

def convert_android_ver(item):
    if pd.isna(item):  # Handles np.nan or float('nan')
        return np.nan
    
    item = str(item).strip()  # Convert to string to handle all cases uniformly
    
    if item == "":
        return np.nan
    elif item == "0":
        return 0
    elif len(item) >= 5:
        return float(item[0:3])
    else:
        return float(item)
```

```python
convert_android_ver("5.0 - 7.1.1")
```

```python
df_copy["Android Ver"] = df_copy["Android Ver"].apply(convert_android_ver) #convert the android ver column to float
df_copy["Android Ver"].dtype
```
